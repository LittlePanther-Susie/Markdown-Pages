```python
import matplotlib.pyplot as plt
import numpy as np
```

### Very Basic

```python
x = np.linspace(0, 10, 100)
y = np.sin(2*x)

fig, ax = plt.subplots(figsize=(5, 4))
ax.plot(x, y, label='sin(t)', linestyle='-', linewidth=2, alpha=0.5, color='blue', marker='P', markersize=10, markerfacecolor='red', markeredgecolor='none')
ax.plot(x, y*0.5, label='0.5sin(t)', linestyle='--', linewidth=1, alpha=0.5, color='green')

ax.set_title('Sine Function', fontsize=14)
ax.set_xlabel('t (sec)', fontsize=12)
ax.set_ylabel('sin(t)', fontsize=12)
ax.legend(loc='best', fontsize=10)
plt.show()
```

### Grid line and tick

```python
x = np.linspace(0, 10, 100)
y = np.sin(2*x)

fig, ax = plt.subplots(figsize=(5, 4))
ax.plot(x, 2*y, label='2*sin(t)', linestyle='--', linewidth=1, alpha=0.5, color='blue', marker='o', markersize=5, markerfacecolor='red', markeredgecolor='none')
ax.scatter(x, y, label='sin(t)', alpha=0.5, color='green')

ax.set(xlim=(0, 11), xticks=np.arange(0, 10.1),
       ylim=(-3.1, 3.1), yticks=np.arange(-3, 3.1, 0.5))
ax.set_title('Sine Function')
ax.set_xlabel('t (sec)')
ax.set_ylabel('sin(t)')
ax.legend(loc='lower right')

# ax.grid()
ax.yaxis.grid(True)
ax.yaxis.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=0.9)
# ax.minorticks_on()
ax.tick_params(which='minor', bottom=False, left=False)

# Axis square
ax.set_aspect('equal', adjustable='box')
# plt.axis('square')

# For all figures, change text size
SMALL_SIZE = 10
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

fig.suptitle('SupTitle', fontsize=20)
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE, labelsize=MEDIUM_SIZE)     # fontsize of the axes title, fontsize of the x and y labels
# plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title. It doesn't work.
plt.show()
```

### Subplots
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(8, 3))
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)

# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)

# # Share a X axis with each column of subplots
# plt.subplots(2, 2, sharex='col')

# # Share a Y axis with each row of subplots
# plt.subplots(2, 2, sharey='row')

# # Share both X and Y axes with all subplots
# plt.subplots(2, 2, sharex='all', sharey='all')

# # Note that this is the same as
# plt.subplots(2, 2, sharex=True, sharey=True)

# # Create figure number 10 with a single subplot
# # and clears it if it already exists.
# fig, ax = plt.subplots(num=10, clear=True)
plt.show()
```
